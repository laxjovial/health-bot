```python
## Preprocess the data and build the model
### Do the first splitting into features and target variable

X = train.drop("prognosis", axis = 1) # features
y = train["prognosis"] # target variable
```

```python
print(y)
## Lets do further preprocessing by encoding our target variable into its numeric representation to ensure a seemless classification
label_encoder = LabelEncoder() # initialize encoder
Y = label_encoder.fit_transform(y)
print(Y)
```

```python
### Do the second splitting for training and testing the model.
##### THe split is in thw ratio of 70:30, 70% to train the algorithm and 30% to test the model
X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size = 0.3, random_state = random_state)
```

```python
## LEts build our model 
#1. Algorithm selection inside a dictionary
## I will be using clusters of classifiers to see the best performing  model
# SVC, Ensembles, Kneighbors, Naive_bayes

training_algorithms = {
    "SVC":SVC(kernel= "linear"),
    "RandomForest": RandomForestClassifier(n_estimators = 100, random_state =random_state),
    "GradientBoosting": GradientBoostingClassifier(n_estimators = 100, random_state =random_state),
    "KNeigbors":KNeighborsClassifier(n_neighbors = 5),
    "MultinormialNB": MultinomialNB()}
```

```python
# Loop all the algorithms together
for algorithm_names, training_algorithm in training_algorithms.items():
    print(algorithm_names,":",training_algorithm)
```

```python
# LEts train our algorithm with our data to get out model
for algorithm_names, training_algorithm in training_algorithms.items():
    models = training_algorithm.fit(X_train,y_train)#.....train
    predictions = models.predict(X_test) #................predict
    accuracy = accuracy_score(y_test,predictions) #......Test
    cm = confusion_matrix(y_test,predictions)#.Evaluate

    #print results
    print(f"{algorithm_names} accuracy: {accuracy}")
    print(f"{algorithm_names} confusion_matrix: ")
    print(np.array2string(cm,separator =','))
    print("_______________________________________________")
    
```

**CHOOSING A MODEL**

MultinomialNB
I will be using Naive_Bayes for my prediction, the reasons are listed below:

Speed:
Training: Very fast, as it computes simple probabilities based on the training data.
Prediction: Very fast, as it involves basic probability calculations.
Suitability: Ideal for text data or categorical features, especially when features are independent.

```python
# TRAINING A SINGLE MODEL WITH NAIVE_BAYES

# Define and train the model
algorithm = MultinomialNB()
model = algorithm.fit(X_train, y_train)
predicted_y = model.predict(X_test)


# Calculate metrics
accuracy = accuracy_score(y_test, predicted_y)
precision = precision_score(y_test, predicted_y, average='weighted')  
recall = recall_score(y_test, predicted_y, average='weighted')  # Sensitivity
f1 = f1_score(y_test, predicted_y, average='weighted')

# Confusion matrix for multi-class classification
conf_matrix = confusion_matrix(y_test, predicted_y)

# Print results
print("Predicted labels: ", predicted_y)
print("True labels: ", y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision (Weighted): {precision:.2f}")
print(f"Recall (Weighted): {recall:.2f}")
print(f"F1-Score (Weighted): {f1:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)

# Generate and print a full classification report
print("\nClassification Report:")
print(classification_report(y_test, predicted_y))
```

```python
# Dictionary comprehension
# Convert the column names of the features to  to a dictionary usinng index as values while the column name becomes the key
symptoms_dict = {column: index + 0 for index, column in enumerate(X.columns)}

print(symptoms_dict)
```

```python
# Dictioanry comprehension...
# Create a dictionary  for the target variable using  both the encoded and the target variable and keeping the uniques values for the dictioanry
disease_dict = {encoded: original for encoded, original in zip(Y, label_encoder.inverse_transform(Y))}

print("disease:", disease_dict)
```

```python
# Creating a prediction function
diseases_list = list(y)
```

```python
diseases_list
```

```python
# Recommender Engine
def predict_ailment(patient_symptoms):
    input_vector = np.zeros(len(symptoms_dict))
    for item in patient_symptoms:
        input_vector[symptoms_dict[item]] = 1
    return diseases_list[my_model.predict([input_vector])[0]]
```

```python
# Added functionalties
def include_features(predicted_disease):
    description = desc[desc["Disease"] ==predicted_disease]["Description"]
    description = "".join([dr for dr in description])
    
    precaution = prec[prec["Disease"]==predicted_disease][["Precaution_1","Precaution_2","Precaution_3","Precaution_4"]]
    precaution = [col for col in precaution.values]

    medication = medic[medic["Disease"]== predicted_disease]["Medication"]
    medication = [med for med in medication.values]

    diets = diet[diet["Disease"] == predicted_disease]["Diet"]
    diets = [di for di in diets.values]

    workout = workouts[workouts["disease"] == predicted_disease]["workout"]
    workout = [wk for wk in workout.values]
    
    return description,precaution,medication,diets,workout
```

```python
from tabulate import tabulate

def generate_formated_report(predicted_disease, description, precaution, medication, diets, workout):
   
    # Prepare rows for the report
    rows = [
        ["\033[1mPredicted Disease\033[0m", predicted_disease],
        ["\033[1mDescription\033[0m", description],
        ["\033[1mPrecautions\033[0m", "\n".join(f"{i+1}. {p}" for i, p in enumerate(precaution[0]))],
        ["\033[1mMedications\033[0m", "\n".join(f" {m}" for m in medication)],
        ["\033[1mDiets\033[0m", "\n".join(f" {d}" for d in diets)],
        ["\033[1mWorkouts\033[0m", "\n".join(f"{i+1}. {w}" for i, w in enumerate(workout))]
    ]

    # Generate the table with `grid` for straight vertical and horizontal lines
    table = tabulate(
        rows,
        headers=["\033[1mPredictions\033[0m", "\033[1mDetails\033[0m"],
        tablefmt="grid"
    )

    # Calculate the length of the table and the separator line length
    table_length = len(table.splitlines()[0])  # Length of the first row in the table

    # Create the header and footer lines with the same length as the table
    header_footer_line = "━" * table_length

    # Print the styled report
    print(header_footer_line)
    print("{:^70}".format("\033[1mPREDICTION AND RECOMMENDATION REPORT\033[0m"))
    print(header_footer_line)
    print(table)
    print(header_footer_line)
    print("{:^70}".format("\033[1mHealth is Wealth!\033[0m"))
    print(header_footer_line)
```
